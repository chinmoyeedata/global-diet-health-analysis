# Distribution Analysis
### The histograms examine the distribution of each numerical variable and identify potential skewness, outliers and overall pattern of the datasets.

## Key Insights:
* Most food consumption variables are right-skewed. This indicates that most countries have low to moderate consumption, while a few consume much larger amounts.
* GDP per Capita is highly right-skewed. This indicates that most countries have lower incomes, while a few countries have very high incomes.
* Life expectancy is fairly consistent, with most countries living between 70–82 years.
* Obesity and Diabetes Prevalence are concentrated at lower values but a few countries have much higher rates.
* Some variables, like GDP per Capita, Milk, Fruits, Vegetables, Sugar and Obesity, contain a few high-value outliers.

```python
correlation = data.select_dtypes(include="number").corr()
correlation
```

```python
import seaborn as sns
plt.figure(figsize=(13, 9))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, center=0, annot_kws={"size":9})
plt.title("Correlation Heatmap", fontsize=16, fontweight="bold")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

# Correlation Analysis
### The correlation heatmap examines the relationship between dietary habits, GDP per capita, life expectancy, diabetes prevalence and obesity across different countries.
## Key Insights:
* Egg consumption has a strong positive correlation with Life Expectancy (r = 0.66).
* Milk consumption is also strongly positively correlated with Life Expectancy (r = 0.64).
* Meat consumption shows a positive correlation with both Life Expectancy (r = 0.58) and Obesity (r = 0.62).
* Sugar consumption has a moderate positive correlation with Obesity (r = 0.44) but a very weak correlation with Diabetes Prevalence (r = 0.07).
* Obesity and Diabetes Prevalence are moderately positively correlated (r = 0.58), suggesting that countries with higher obesity rates tend to have higher diabetes prevalence.
* GDP per Capita has the strongest positive correlation with Life Expectancy (r = 0.68), indicating that wealthier countries generally have longer life expectancy.

# Scatter PLot 1: GDP per Capita vs Life Expectancy

```python
plt.figure(figsize=(8,5))
plt.scatter(data["GDP per Capita (US$)"], data["Life Expectancy (Years)"], alpha=0.7)
plt.grid(alpha=0.3)
plt.xlabel("GDP per Capita (US$)")
plt.ylabel("Life Expectancy (Years)")
plt.title("GDP per Capita vs Life Expectancy")
plt.show()
```

## Key Insights:
* Countries with higher GDP per capita generally have higher life expectancy.
* The increase in life expectancy is much faster at lower income levels.
* After a certain income level, life expectancy improves only slightly even if GDP continues to increase.
* This suggests that economic growth is important for improving health, especially in low and middle income countries.
